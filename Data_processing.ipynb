```python
## Read the excel file using Pandas
import pandas as pd
df=pd.read_excel(r'D:\ARUL\Data_Science\Projrect-1\traffic_stops.xlsx')
df
```

```python
##To reduce the number of columns, I have combined the date and time into a single column
df['Stop_datetime'] = pd.to_datetime(df['stop_date'].astype(str) + ' ' + df['stop_time'].astype(str))
df.head()
```

```python
##Drop unwanted columns
df.drop(['driver_age_raw','violation_raw'],axis=1,inplace=True)
df.head()
```

```python
##Checking null values in my data
df.isnull().sum()
```

```python
## fill Nan value into No Search
df['search_type'] = df['search_type'].fillna('No Search')
df.head(10)
```

```python
## View the Columns in the excel
df.columns
```

```python
##Connect the MYSQL to create the table for my data structure
import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="Arul@123",
  auth_plugin='mysql_native_password'
)

mycursor = mydb.cursor()
mycursor.execute('use police;')

create_table_query="""
CREATE TABLE traffic_stops (
    Pk_id INT AUTO_INCREMENT PRIMARY KEY,
    country_name VARCHAR(100) NOT NULL,
    driver_gender CHAR(1),
    driver_age INT CHECK (driver_age >= 0),
    driver_race VARCHAR(50),
    violation VARCHAR(100),
    search_conducted BOOLEAN,
    search_type VARCHAR(50),
    stop_outcome VARCHAR(50),
    is_arrested BOOLEAN,
    stop_duration INT CHECK (stop_duration >= 0),
    drugs_related_stop BOOLEAN,
    vehicle_number VARCHAR(50),
    Stop_datetime DATETIME
);
"""

mycursor.execute(create_table_query)
mydb.commit()
mycursor.close()
mydb.close()

print("Table 'traffic_stops' created successfully!")
```

```python
##Connect the MYSQL insert the excel data into my table

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="Arul@123",
  auth_plugin='mysql_native_password'
)

mycursor = mydb.cursor()
mycursor.execute('use police;')


insert_query = """
INSERT INTO traffic_stops (
    country_name, driver_gender, driver_age, driver_race, violation,
    search_conducted, search_type, stop_outcome, is_arrested,
    stop_duration, drugs_related_stop, vehicle_number, Stop_datetime
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

data_tuples = [
    (
        row['country_name'],
        row['driver_gender'],
        row['driver_age'],
        row['driver_race'],
        row['violation'],
        row['search_conducted'],
        row['search_type'],
        row['stop_outcome'],
        row['is_arrested'],
        row['stop_duration'],
        row['drugs_related_stop'],
        row['vehicle_number'],
        row['Stop_datetime'].to_pydatetime() if pd.notna(row['Stop_datetime']) else None
    )
    for _, row in df.iterrows()
]

mycursor.executemany(insert_query, data_tuples)

mydb.commit()
mycursor.close()
mydb.close()
```
